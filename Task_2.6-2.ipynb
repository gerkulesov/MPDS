```python
# Общая часть

import numpy as np # математика
import matplotlib.pyplot as plt # графики

# Апроксимируемая функция
def f(x):
    return np.sin(x/5.) * np.exp(x/10.) + 5 * np.exp(-x/2.)

# Апроксимирующая функция
# coefficients -- список коэффициентов w_0, w_1, ..., w_n
def aprox_f(x, coefficients):
    return sum([coefficients[i] * x**i for i in range(len(coefficients))])
```

```python
# Часть 1: апроксимация линейной функцией

deg = 1 # степень функции
x_list = [1, 15] # фиксированные точки

A = np.array([[x**n for n in range(deg+1)] for x in x_list]) # матрица коэффициентов, степеней x
B = np.array([f(x) for x in x_list]) # вектор значений
X = np.linalg.solve(A, B) # искомые коэффициенты w_0, w_1

print 'Матрица коэффициентов:\n', A
print '\nВектор значений:\n', B
print '\nРешение СЛАУ:\n', X

dx = 0.01 # шаг для построения графика
x = np.arange(1, 15, dx) # точки, в которых строится график
y_original = [f(t) for t in x] # значения апроксимируемой функции
y_aproxima = [aprox_f(t, X) for t in x] # значения апроксимирующей функции

plt.plot(x, y_original, label='original') # построение графика апроксимируемой функции
plt.plot(x, y_aproxima, label='aproximated') # построение графика апроксимирующей функции
[plt.vlines(x, 0, f(x), linestyles='dotted') for x in x_list] # построение вертикальных маркеров в фиксированных точках
plt.xticks(range(1, 16)) # поправляем подписи чисел на оси x
plt.legend() # добавляем легенду на график
plt.show() # строим график
```

```python
# Часть 2: апроксимация квадратным трёхчленом

deg = 2
x_list = [1, 8, 15]

A = np.array([[x**n for n in range(deg+1)] for x in x_list])
B = np.array([f(x) for x in x_list])
X = np.linalg.solve(A, B) # искомые коэффициенты w_0, w_1, w_2

print 'Матрица коэффициентов:\n', A
print '\nВектор значений:\n', B
print '\nРешение СЛАУ:\n', X

dx = 0.01
x = np.arange(1, 15, dx)
y_original = [f(t) for t in x]
y_aproxima = [aprox_f(t, X) for t in x]

plt.plot(x, y_original, label='original')
plt.plot(x, y_aproxima, label='aproximated')
[plt.vlines(x, 0, f(x), linestyles='dotted') for x in x_list]
plt.xticks(range(1, 16))
plt.legend()
plt.show()
```
